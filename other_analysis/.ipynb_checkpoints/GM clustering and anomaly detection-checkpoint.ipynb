## Silhouette Score: determining best number of GM components

```python
from sklearn import metrics
from sklearn.mixture import GaussianMixture
```

```python
# # GMM plot function
# def plot_gmm(gmm, X, label=True, ax=None):
#     ax = ax or plt.gca()
#     labels = gmm.fit(X).predict(X)
#     if label:
#         ax.scatter(X[:, 0], X[:, 1], c=labels, s=40, cmap='viridis', zorder=2)
#     else:
#         ax.scatter(X[:, 0], X[:, 1], s=40, zorder=2)
    
#     w_factor = 0.2 / gmm.weights_.max()
#     for pos, covar, w in zip(gmm.means_, gmm.covariances_, gmm.weights_):
#         draw_ellipse(pos, covar, alpha=w * w_factor)
#     plt.title("GMM with %d components"%len(gmm.means_), fontsize=(20))
#     plt.xlabel("U.A.")
#     plt.ylabel("U.A.")
```

```python
def SelBest(arr:list, X:int)->list:
    '''
    returns the set of X configurations with shorter distance
    '''
    dx=np.argsort(arr)[:X]
    return arr[dx]
```

```python
n_clusters=np.arange(2, 10)
sils=[]
sils_err=[]
iterations=20
for n in n_clusters:
    tmp_sil=[]
    for _ in range(iterations):
        gmm=GaussianMixture(n, n_init=2).fit(cluster_data) 
        labels=gmm.predict(cluster_data)
        sil=metrics.silhouette_score(cluster_data, labels, metric='euclidean')
        tmp_sil.append(sil)
    val=np.mean(SelBest(np.array(tmp_sil), int(iterations/5)))
    err=np.std(tmp_sil)
    sils.append(val)
    sils_err.append(err)
    
plt.errorbar(n_clusters, sils, yerr=sils_err)
plt.title("Silhouette Scores", fontsize=20)
plt.xticks(n_clusters)
plt.xlabel("N. of clusters")
plt.ylabel("Score")
```

## Gaussian mixture: 4 components

```python
GM=GaussianMixture(n_components=4, random_state=999) #note: account for differences b/t rand states
GM.fit(cluster_data)
```

```python
GM_LL = pd.DataFrame()
GM_LL['LL'] = GM.score_samples(cluster_data)
GM_LL['Neighborhood'] = rim['Neighborhood']
GM_LL['cluster'] = GM.predict(cluster_data)
```

```python
# calcualte probability of each neighborhood belonging to each cluster
probs = GM.predict_proba(cluster_data)
```

```python
probs = pd.DataFrame(probs)
probs.columns = ['prob_0', 'prob_1', 'prob_2', 'prob_3']
```

```python
GM_results = GM_LL.merge(probs, on=probs.index)
GM_results = GM_results.drop(columns='key_0')
```

```python
GM_results = GM_results.sort_values(by='Neighborhood')
```

## Summary stats on input features per cluster assignment

```python
rim['cluster'] = GM_results['cluster']
rim
```

```python
neighborhoods0 = []
neighborhoods1 = []
neighborhoods2 = []
neighborhoods3 = []

for i in range(0, len(rim)):
    if rim['cluster'][i] == 0:
        neighborhoods0.append(rim['Neighborhood'][i])
    if rim['cluster'][i] == 1:
        neighborhoods1.append(rim['Neighborhood'][i])
    if rim['cluster'][i] == 2:
        neighborhoods2.append(rim['Neighborhood'][i])
    if rim['cluster'][i] == 3:
        neighborhoods3.append(rim['Neighborhood'][i])
   
```

```python
cluster_summary = pd.DataFrame()

cluster_summary['BusinessLoss_mean'] = rim.groupby('cluster')['BusinessLossMetric'].mean()
cluster_summary['StaffCancellation_mean'] = rim.groupby('cluster')['StaffCancellationMetric'].mean()
cluster_summary['VenuesPer1k_mean'] = rim.groupby('cluster')['VenuesPer1kMetric'].mean()
cluster_summary['OLS_PCT_mean'] = rim.groupby('cluster')['OLS_PCT_Metric'].mean()
cluster_summary['VarietyMetric_mean'] = rim.groupby('cluster')['VarietyMetric'].mean()
```

```python
cluster_summary['neighborhoods'] = pd.Series()
cluster_summary['neighborhoods'] = cluster_summary['neighborhoods'].astype('object')
cluster_summary.at[0,'neighborhoods'] = neighborhoods0
cluster_summary.at[1,'neighborhoods'] = neighborhoods1
cluster_summary.at[2,'neighborhoods'] = neighborhoods2
cluster_summary.at[3,'neighborhoods'] = neighborhoods3
cluster_summary.reset_index(inplace=True)
cluster_summary
```
