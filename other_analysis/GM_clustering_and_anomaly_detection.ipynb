# Gaussian Mixture Clustering

### GM clustering of risk features included in Neighborhood Profile Dashboard

Input files:  
1. Neighborhood_Profile_Dashboard.csv, containing risk features displayed in dashboard of Neighborhood Profile
2. ZipsDissolve.shp (shape file of NYC zip codes outlines)
3. boro_new.shp (shape file of NYC borough outlines)

Output files:  Image (heatmap of neighborhood cluster assignments) 

```python
import numpy as np
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
import warnings
warnings.filterwarnings("ignore")
```

## Data Preprocessing

```python
# risk input metrics
rim = pd.read_csv('Neighborhood Profile Dashboard.csv')
```

```python
rim.shape
```

```python
rim['Variety'] = rim['Variety']/100
rim.head()
```

```python
cluster_data = rim.iloc[:,1:6]
```

```python
cluster_data.shape
```

## Selecting Number of Clusters with BIC Gradient

```python
def SelBest(arr:list, X:int)->list:
    '''
    returns the set of X configurations with shorter distance
    '''
    dx=np.argsort(arr)[:X]
    return arr[dx]
```

```python
n_clusters=np.arange(2, 10)
bics=[]
bics_err=[]
iterations=20
for n in n_clusters:
    tmp_bic=[]
    for _ in range(iterations):
        gmm=GaussianMixture(n, n_init=2).fit(cluster_data) 
        tmp_bic.append(gmm.bic(cluster_data))
    val=np.mean(SelBest(np.array(tmp_bic), int(iterations/5)))
    err=np.std(tmp_bic)
    bics.append(val)
    bics_err.append(err)
```

```python
plt.errorbar(n_clusters,bics, yerr=bics_err, label='BIC')
plt.title("BIC Scores", fontsize=20)
plt.xticks(n_clusters)
plt.xlabel("N. of clusters")
plt.ylabel("Score")
plt.legend()
```

```python
plt.errorbar(n_clusters, np.gradient(bics), yerr=bics_err, label='BIC')
plt.title("Gradient of BIC Scores", fontsize=20)
plt.xticks(n_clusters)
plt.xlabel("N. of clusters")
plt.ylabel("grad(BIC)")
plt.legend()
```

Using the BIC gradient method, we see that 4 is the optimal number of clusters because the gradient starts leveling off at 5 clusters.

## Gaussian mixture: 4 clusters

```python
# iterate over several random states to obtain most stable cluster configuration
random_states = [39, 99, 14, 56, 71, 3, 7, 33, 25, 82]
GM_Results = pd.DataFrame()
for i in range(1, len(random_states)+1):
    GM=GaussianMixture(n_components=4, random_state = random_states[i-1]) 
    GM.fit(cluster_data)
    GM_Results['Neighborhood'] = rim['Neighborhood']
    GM_Results['cluster{}'.format(i)] = GM.predict(cluster_data)
    GM_Results.sort_values(by='Neighborhood')
```

```python
GM_Results['Aligned_Mode'] = GM_Results['cluster1']
# The mode of the ten cluster results, once the cluster numbers were realigned, was equivalent to cluster1's result.
# Please see GM Modal Clustering csv for more details.
```

```python
GM_Results
```

## Summary stats on input features per cluster assignment

```python
rim['cluster'] = GM_Results['Aligned_Mode']
rim
```

```python
# create list of neighborhood names in each cluster
neighborhoods0 = []
neighborhoods1 = []
neighborhoods2 = []
neighborhoods3 = []

for i in range(0, len(rim)):
    if rim['cluster'][i] == 0:
        neighborhoods0.append(rim['Neighborhood'][i])
    if rim['cluster'][i] == 1:
        neighborhoods1.append(rim['Neighborhood'][i])
    if rim['cluster'][i] == 2:
        neighborhoods2.append(rim['Neighborhood'][i])
    if rim['cluster'][i] == 3:
        neighborhoods3.append(rim['Neighborhood'][i])
```

```python
cluster_summary = pd.DataFrame()
cluster_summary['BusinessLoss_mean'] = rim.groupby('cluster')['BusinessLoss'].mean()
cluster_summary['StaffCancellation_mean'] = rim.groupby('cluster')['StaffCancellation'].mean()
cluster_summary['VenuesPer1k_mean'] = rim.groupby('cluster')['Venuesper1K'].mean()
cluster_summary['OLS_PCT_mean'] = rim.groupby('cluster')['OLS_PCT'].mean()
cluster_summary['VarietyMetric_mean'] = rim.groupby('cluster')['Variety'].mean()
```

```python
cluster_summary['neighborhoods'] = pd.Series()
cluster_summary['neighborhoods'] = cluster_summary['neighborhoods'].astype('object')
cluster_summary.at[0,'neighborhoods'] = neighborhoods0
cluster_summary.at[1,'neighborhoods'] = neighborhoods1
cluster_summary.at[2,'neighborhoods'] = neighborhoods2
cluster_summary.at[3,'neighborhoods'] = neighborhoods3
cluster_summary.reset_index(inplace=True)
cluster_summary
```
